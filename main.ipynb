```python
import pandas as pd
data=pd.read_csv("ecoProj.csv")
```

```python
data[data['sku_id']==216233]
```

```python
#group by sku_id
grouped=data.groupby(['sku_id'])
```

```python
for group in grouped:
    print(group)
```

```python
#for each group find the correlation between the total_price and units_sold columns
corr_list=[]
for group in grouped:
    print(group[0]," :")
    print(group[1]['total_price'].corr(group[1]['units_sold']))
    corr_list.append(group[1]['total_price'].corr(group[1]['units_sold']))
```

```python
print(min(corr_list))
#find the sku_id with the minimum correlation
print(corr_list.index(min(corr_list)))
```

```python
corr_list2=[]
for group in grouped:
    print(group[0]," :")
    print(group[1]['base_price'].corr(group[1]['units_sold']))
    corr_list2.append(group[1]['base_price'].corr(group[1]['units_sold']))
```

```python
print(min(corr_list2))
#find the sku_id with the minimum correlation
print(corr_list.index(min(corr_list2)))
```
